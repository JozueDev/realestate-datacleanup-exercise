#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

```python
import pandas as pd  

# Cargar el dataset  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  

# Crear una nueva columna para el precio por metro cuadrado  
ds['pps'] = ds['price'] / ds['surface']   

# Filtrar los datos para cada municipio  
valdemorillo = ds[ds['address'] == 'Valdemorillo']  
galapagar = ds[ds['address'] == 'Galapagar']  

# Calcular el precio promedio por metro cuadrado para cada municipio  
precio_promedio_pps_valdemorillo = valdemorillo['pps'].mean()  
precio_promedio_pps_galapagar = galapagar['pps'].mean()  

# Imprimir los resultados  
print(f"Precio promedio por m² en Valdemorillo: {precio_promedio_pps_valdemorillo:.2f} €")  
print(f"Precio promedio por m² en Galapagar: {precio_promedio_pps_galapagar:.2f} €")  

# Conclusión  
if precio_promedio_pps_valdemorillo == precio_promedio_pps_galapagar:  
    print("Los precios promedios por metro cuadrado de Valdemorillo y Galapagar son los mismos.")  
else:  
    print("Los precios promedios por metro cuadrado de Valdemorillo y Galapagar son diferentes.")  
```

#### Ejercicio 16. Calcula la media y la varianza de muestra para las siguientes variables: precio, habitaciones, superficie y baños (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)

Imprime ambos valores por cada variable.

```python
import pandas as pd  

# Cargamos el data set  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  

# Definimos las poblaciones de interés  
poblaciones_interes = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']  

# Filtramos el DataFrame  
subconjunto = ds[ds['level5'].isin(poblaciones_interes)]  

# Calcular y mostrar la media y la varianza de muestra para cada variable  
variables = ['price', 'rooms', 'surface', 'bathrooms']  
resultados = {}  

for variable in variables:  
    media = subconjunto[variable].mean()  
    varianza = subconjunto[variable].var()  # Varianza de muestra  
    resultados[variable] = {'media': media, 'varianza': varianza}  

# Imprimir los resultados  
for variable, stats in resultados.items():  
    print(f"Variable: {variable}")  
    print(f"  Media: {stats['media']}")  
    print(f"  Varianza: {stats['varianza']}\n")  
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Cargamos el data set  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  

# Definimos las poblaciones de interés  
poblaciones_interes = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']  

# Filtramos el DataFrame  
subconjunto = ds[ds['level5'].isin(poblaciones_interes)].copy()  # Hacemos una copia  

# Normalizamos los precios para cada población  
subconjunto.loc[:, 'precio_normalizado'] = subconjunto.groupby('level5')['price'].transform(  
    lambda x: (x - x.min()) / (x.max() - x.min())  
)  

# Trazar histogramas  
plt.figure(figsize=(10, 6))  

for poblacion in poblaciones_interes:  
    plt.hist(  
        subconjunto[subconjunto['level5'] == poblacion]['precio_normalizado'],  
        bins=10,  
        alpha=0.5,  
        label=poblacion  
    )  

# Configurar el gráfico  
plt.title('Histogramas de Precios Normalizados por Población')  
plt.xlabel('Precio Normalizado')  
plt.ylabel('Frecuencia')  
plt.legend(title='Poblaciones')  
plt.grid(axis='y')  

# Mostrar el gráfico  
plt.tight_layout()  
plt.show()  
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
import pandas as pd  

# Cargamos el data set  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  

# Definimos las poblaciones de interés  
poblaciones_interes = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']  

# Filtramos el DataFrame y creamos una copia  
subconjunto = ds[ds['level5'].isin(poblaciones_interes)].copy()  

# Calcular el precio por metro cuadrado (pps) 
subconjunto.loc[:, 'pps'] = subconjunto['price'] / subconjunto['surface']  

# Filtrar solo Getafe y Alcorcón  
pp_municipios = subconjunto[subconjunto['level5'].isin(['Getafe', 'Alcorcón'])]  

# Calcular la media y la varianza de pps para cada municipio  
resultados_pps = pp_municipios.groupby('level5')['pps'].agg(['mean', 'var']).reset_index()  

# Imprimir los resultados  
print(resultados_pps)  
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Cargamos el data set  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  

# Definimos las poblaciones de interés  
poblaciones_interes = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón', 'Móstoles', 'Torrejón de Ardoz']  

# Filtramos el DataFrame  
subconjunto = ds[ds['level5'].isin(poblaciones_interes)].copy()  

# Seleccionamos cuatro poblaciones diferentes para el gráfico  
poblaciones_a_graficar = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']  

# Crear subgráficos  
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 filas, 2 columnas  

# Crear un diagrama de dispersión para cada población  
for ax, poblacion in zip(axs.flatten(), poblaciones_a_graficar):  
    subset = subconjunto[subconjunto['level5'] == poblacion]  
    ax.scatter(subset['surface'], subset['price'], alpha=0.7)  
    ax.set_title(f'Diagrama de Dispersión: {poblacion}')  
    ax.set_xlabel('Superficie (m2)')  
    ax.set_ylabel('Precio (€)')  
    ax.grid(True)  

# Ajustar el layout  
plt.tight_layout()  
plt.suptitle('Diagrama de Dispersión entre Precio y Superficie por Población', fontsize=16, y=1.02)  
plt.show()  
```

```python
import pandas as pd  
from ipyleaflet import Map, basemaps, Marker, Icon  


data = {  
    'level5': ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'],  
    'lat': [40.3080, 40.3625, 40.3085, 40.3450],  
    'lon': [-3.8043, -3.8044, -3.7333, -3.8280]  
}  

# Creando un DataFrame ficticio  
ds = pd.DataFrame(data)  

# Definimos las poblaciones de interés  
poblaciones_interes = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']  

# Filtrar el DataFrame para obtener solo las poblaciones de interés  
subconjunto = ds[ds['level5'].isin(poblaciones_interes)][['level5', 'lat', 'lon']].copy()  

# Crear un diccionario de coordenadas por población  
coordenadas = {}  
for poblacion in poblaciones_interes:  
    sub = subconjunto[subconjunto['level5'] == poblacion]  
    if sub.empty:  
        print(f"No hay datos para la población: {poblacion}")  
        continue  
    
    coordenadas[poblacion] = {  
        'coordenadas': sub[['lat', 'lon']].values.tolist(),  
        'color': {  
            'Fuenlabrada': 'red',  
            'Leganés': 'blue',  
            'Getafe': 'green',  
            'Alcorcón': 'orange'  
        }[poblacion]  
    }  

# Mapa centrado en Madrid  
mapa = Map(center=(40.35, -3.8), zoom=11.5, min_zoom=1, max_zoom=20,   
            basemap=basemaps.OpenStreetMap.Mapnik)  

# Agregar los marcadores de las coordenadas al mapa  
for poblacion, datos in coordenadas.items():  
    for coord in datos['coordenadas']:  
        lat, lon = coord  
        
        # Crear un icono personalizado  
        marcador_icono = Icon(icon_color=datos['color'], icon_size=(20, 20))  
        
        # Crear el marcador con el icono personalizado  
        marker = Marker(location=(lat, lon), title=poblacion, draggable=False, icon=marcador_icono)  
        
        # Agregar el marcador al mapa  
        mapa.add_layer(marker)  

# Mostrar el mapa  
mapa  
```
