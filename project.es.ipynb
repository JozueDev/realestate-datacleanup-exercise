#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
import pandas as pd  

# Cargar el dataset  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  
 #Filtrar datos para cada municipio
valdemorillo = ds[ds['address'] == 'Valdemorillo']
galapagar = ds[ds['address'] == 'Galapagar']
#desplegar el promedio de precios de cada direccion
valdemorillo_prom = valdemorillo['price'].mean()
galapagar_prom = galapagar['price'].mean()
#imprimir los resultados
print(f"Precio promedio de Valdemorillo: {valdemorillo_prom:.2f} € ")
print(f"Precio promedio de Galapagar: {galapagar_prom:.2f} € ")
#conclusion
if valdemorillo_prom == galapagar_prom:
    print("Los precios promedios de Valdemorillo y Galapagar son los mismos")
else:
    print("Los precios promedios de Valdemorillo y Galapagar son diferentes")
```

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
import pandas as pd  

# Cargar el dataset  
ds = pd.read_csv('assets/real_estate.csv', sep=';')  

# Crear una nueva columna para el precio por metro cuadrado  
ds['pps'] = ds['price'] / ds['surface']   

# Filtrar los datos para cada municipio  
valdemorillo = ds[ds['address'] == 'Valdemorillo']  
galapagar = ds[ds['address'] == 'Galapagar']  

# Calcular el precio promedio por metro cuadrado para cada municipio  
precio_promedio_pps_valdemorillo = valdemorillo['pps'].mean()  
precio_promedio_pps_galapagar = galapagar['pps'].mean()  

# Imprimir los resultados  
print(f"Precio promedio por m² en Valdemorillo: {precio_promedio_pps_valdemorillo:.2f} €")  
print(f"Precio promedio por m² en Galapagar: {precio_promedio_pps_galapagar:.2f} €")  

# Conclusión  
if precio_promedio_pps_valdemorillo == precio_promedio_pps_galapagar:  
    print("Los precios promedios por metro cuadrado de Valdemorillo y Galapagar son los mismos.")  
else:  
    print("Los precios promedios por metro cuadrado de Valdemorillo y Galapagar son diferentes.")  
```

#### Ejercicio 11. Analiza la relación entre la superficie y el precio de las casas. (★★☆)

```python
import pandas as pd
import matplotlib.pyplot as plt

#cargar dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')
#crear scatter plot
plt.figure(figsize=(10,6))
plt.scatter(x = ds['surface'], y= ds['price'], alpha=0.6)
#personalizar grafica 
plt.title('Relación entre Superficie (m²) y Precio de las Casas')  
plt.xlabel('Superficie (m²)')  
plt.ylabel('Precio (€)')  
plt.xscale('linear')  # Escala lineal para el eje x  
plt.yscale('linear')  # Escala lineal para el eje y  
plt.grid(True)  

#mostrar grafica
plt.show()
```

#### Ejercicio 13. ¿Cuál es la población (columna level5) que contiene la mayor cantidad de casas?(★★☆)

Imprima la población y el número de casas.

```python
import pandas as pd


#cargar dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')
#agrupar por la columna 'level5' y contar el numero de casas
conteo = ds.groupby('level5').size()  
#filtrar la poblacion con mayor numeor de casas
poblacion_max = conteo.idxmax()
conteo_casas = conteo.max()
#imprimir resultado
print(f"La poblacion con la mayor cantidad de casas es: {poblacion_max} con {conteo_casas} casas.")
```
